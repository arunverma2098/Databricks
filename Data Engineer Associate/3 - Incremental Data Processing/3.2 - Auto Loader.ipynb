```python
%sql
USE CATALOG workspace;
USE SCHEMA bookstore_eng_ass;
```

```python
(
    spark.readStream
         .format("cloudFiles")
         .option("cloudFiles.format","parquet")
         .option("cloudFiles.schemaLocation", "/Volumes/workspace/bookstore_eng_ass/checkpoints/orders_checkpoint")
         .load("/Volumes/workspace/bookstore_eng_ass/dataset/orders-raw/")
         .writeStream
         .option("checkpointLocation", "/Volumes/workspace/bookstore_eng_ass/checkpoints/orders_checkpoint")
         .trigger(availableNow=True)
         .table("orders_updates")
         .awaitTermination()
)
```

```python
%sql
select * from orders_updates
```

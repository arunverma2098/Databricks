```python
(
    spark.readStream
            .table("workspace.bookstore_eng_ass.books")
            .createOrReplaceTempView("books_streaming_vw")
)
```

```python
%sql
create or replace temp view author_counts_tmp_vw as (
  select author, count(book_id) as total_books
  from books_streaming_vw
  group by author
)
```

```python
(
    spark.table("author_counts_tmp_vw")
    .writeStream
    .trigger(availableNow=True)
    .outputMode("complete")
    .option("checkpointLocation", "/Volumes/workspace/bookstore_eng_ass/checkpoints/author_counts_checkpoint")
    .table("author_counts")
    .awaitTermination()
)
```

```python
%sql
select * from author_counts
```
